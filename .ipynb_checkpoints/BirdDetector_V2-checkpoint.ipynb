```python
default_resolution = 300
```

```python
def normalize(input):
    normalized_input = input / 255.0
    return normalized_input.astype(np.float16)
```

```python
def resize(imagem):
    # Define a largura e altura desejadas
    largura_desejada, altura_desejada = default_resolution, default_resolution

    # Obtém as dimensões da imagem original
    altura, largura = imagem.shape[:2]

    # Calcula a proporção de redimensionamento
    proporcao_largura = largura_desejada / largura
    proporcao_altura = altura_desejada / altura

    # Escolhe a proporção mínima para garantir que a imagem caiba no novo tamanho
    proporcao = min(proporcao_largura, proporcao_altura)

    # Calcula as novas dimensões da imagem
    nova_largura = int(largura * proporcao)
    nova_altura = int(altura * proporcao)

    # Redimensiona a imagem para as novas dimensões
    imagem_redimensionada = cv2.resize(imagem, (nova_largura, nova_altura))

    # Cria uma imagem em branco com o tamanho desejado
    imagem_final = np.ones((altura_desejada, largura_desejada, 3), dtype=np.uint8) * 255

    # Calcula as coordenadas de início para colar a imagem redimensionada no centro
    x_inicio = (largura_desejada - nova_largura) // 2
    y_inicio = (altura_desejada - nova_altura) // 2

    # Cola a imagem redimensionada no centro da imagem em branco
    imagem_final[y_inicio:y_inicio + nova_altura, x_inicio:x_inicio + nova_largura] = imagem_redimensionada

    return imagem_final
```

```python

# Root directory to scan
root_directory_train = '../BirdsDataset2/Train'
root_directory_test = '../BirdsDataset2/Test'

# Function to list all files and subdirectories
def list_files_and_subdirectories(directory):
    subdirectories = []
    files = []

    for item in os.listdir(directory):
        item_path = os.path.join(directory, item)
        if os.path.isdir(item_path):
            subdirectories.append(item)
        else:
            files.append(item)

    return subdirectories, files

# Initialize lists to store information
class_id = []
file_path = []
labels = []
dataset = []

# Function to process a directory
def process_directory(directory_name, directory_path, dataset_model):
    global current_class  # Mark current_class as a global variable
    subdirectories, files = list_files_and_subdirectories(directory_path)

    # Add information to the lists
    for file in files:
        class_id.append(current_class)
        file_path.append(os.path.join(dataset_model + '/' + directory_name, file))
        labels.append(directory_name)
        dataset.append(dataset_model)

    for subdirectory in subdirectories:
        # Recursively process subdirectories with a new class
        current_class += 1
        process_directory(subdirectory, os.path.join(directory_path, subdirectory), dataset_model)

# Initial class ID
current_class = 0

# Start the process with the root directory
process_directory('', root_directory_train, 'Train')
current_class = 0  # Reset current_class to zero
process_directory('', root_directory_test, 'Test')

# Create a DataFrame using Pandas
data = pd.DataFrame({
    'class id': class_id,
    'filepaths': file_path,
    'labels': labels,
    'data set': dataset
})

# Save the data to a CSV file if needed
data.to_csv('data.csv', index=False)
```

```python
i=0

train_images = []
train_labels = []
for class_id in range(n_class):  # O range inclui até 200
    filtered_images = train_data[train_data['class id'] == class_id]

    n_images_per_class_to_load_int = n_images_per_class_to_load
    
    if n_images_per_class_to_load > len(filtered_images):        
        n_images_per_class_to_load_int = len(filtered_images)
        print(len(filtered_images))
        
    #print('ID ' +str(class_id) +  ' ' + str(len(filtered_images)) + ' '+ str(n_images_per_class_to_load) + ' '+ str(n_images_per_class_to_load_int))

   # print(n_images_per_class_to_load)
    for index, row in islice(filtered_images.iterrows(), n_images_per_class_to_load_int):                  
            path='..\\BirdsDataset2\\' + row['filepaths']    #Creating the path
            image_path = path #os.path.join(path)               #Creating the OS Path to the image        
            image = cv2.imread(image_path)
            
            height, width, channels = image.shape   
            if i < len(train_images):                    
                train_images[i] =  resize(image)                
                train_labels[i] = int(row['class id'])
            else:
                train_images.append( resize(image))   
                train_labels.append(int(row['class id']))
            i+=1
```

```python

#Distribution strategy - This will enable TF to use CPU and GPU
strategy = tf.distribute.MirroredStrategy()
height = default_resolution
width = default_resolution
# Defining a function to create our model
def create_model():
    base_model = keras.applications.ResNet50(weights='imagenet', include_top=False, input_shape=(height, width, channels))

    #Freezing the Resnet50 model
    for layer in base_model.layers:
        layer.trainable = False

    model = keras.Sequential([
        #layers.Rescaling(1./255, input_shape=(height, width, channels)),
        base_model,
        layers.Flatten(),
        layers.Dense(1000, activation='relu'),
        layers.Dense(n_class, activation='softmax')
    ])

    #for layer in model.layers[1].layers:
    #    layer.trainable = False
        
    return model

# Creating the model
with strategy.scope():
    model = create_model()

#Compiling the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
